```python
import pandas as pd
import numpy as np
from collections import Counter
import matplotlib.pyplot as plt
```

```python
# read the file
df = pd.read_csv('C:/users/idaid/Desktop/MA Thesis/data/English_excerpts_multilingual.csv')
df
```

```python
# Calulate frequencies
annotations = ['Summary','Tone', 'Intertextuality', 'Genre','Literary movement', 'Literary form', 'Trope', 
         'Enunciation', 'Narrative arc', 'Speech standard','Active character', 'Quoted character',  'Quoted work', 
         'Fuzzy place','Absolute place', 'Fuzzy time', 'Time setting', 'Diegetic time', 'Absolute time']

statistic_dict = {}
# Iterate through the analysis
for analysis in df.analysis.tolist():
    # Split the annotations
    temp = [text.split(':')[0] for text in analysis.split("'")]
    for t in temp:
        if t not in annotations:
            continue
        # Count frequencies
        if t not in statistic_dict:
            statistic_dict[t] = 1
        else:
            statistic_dict[t] += 1
            
print(statistic_dict)
# Covert to dataframe
df2 = pd.DataFrame([statistic_dict])
# Concatenate df,df2
df3 = pd.concat([df, df2], axis=1)
```

```python
# Sample data
data = {
    'annotation': ['Summary', 'Enunciation','Tone','Genre', 'Intertextuality','Speech standard', 'Literary form',
                   'Active character','Trope', 'Narrative arc'],
    
    'frequency': [15, 15, 15, 13, 9, 14, 13, 3, 11, 12]
}

# Creating a DataFrame
df = pd.DataFrame(data)

# Sorting the DataFrame by frequency in descending order
df_sorted = df.sort_values(by='frequency', ascending=False)

# Creating a horizontal bar chart
plt.figure()
bars = plt.barh(df_sorted['annotation'], df_sorted['frequency'])
plt.xlabel('Frequency')
plt.ylabel('Annotation')

# Adding  the frequency numbers on the bars
for bar in bars:
    plt.text(bar.get_width(), bar.get_y() + bar.get_height()/2, 
             f'{bar.get_width():.0f}', 
             va='center', ha='left')
# Invert y axis to have the largest at the top
plt.gca().invert_yaxis() 
plt.show()
```
