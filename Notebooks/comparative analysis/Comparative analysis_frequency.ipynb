## Calculating frequencies for different LLMs in comparative analysis

This notebook contains codes for calculating frequencies for Brahe-AWQ, Claude-20b, Mixtral-8x22b and Llama-3-70b LLMs in comparative analysis.

```python
import pandas as pd
import numpy as np
from collections import Counter
import matplotlib.pyplot as plt
```

**Brahe-AWQ**

```python
# read the file
df = pd.read_csv('C:/users/idaid/Desktop/MA Thesis/data/comparative analysis/Brahe-AWQ-dataset.csv')
df
```

```python
# Calulate frequencies
annotations = ['Summary','Tone', 'Intertextuality', 'Genre','Literary movement', 'Literary form', 'Trope', 
         'Enunciation', 'Narrative arc', 'Speech standard','Active character', 'Quoted character',  'Quoted work', 
         'Fuzzy place','Absolute place', 'Fuzzy time', 'Time setting', 'Diegetic time', 'Absolute time']

statistic_dict = {}
# Iterate through the analysis
for analysis in df.analysis.tolist():
    # Split the annotations
    temp = [text.split(':')[0] for text in analysis.split("\n")]
    for t in temp:
        if t not in annotations:
            continue
        # Count frequencies
        if t not in statistic_dict:
            statistic_dict[t] = 1
        else:
            statistic_dict[t] += 1
            
print(statistic_dict)
# Convert to dataframe
df2 = pd.DataFrame([statistic_dict])
# Concatenate df, df2
df3 = pd.concat([df, df2], axis=1)
```

```python
# Sample data
data = {
    'annotation': ['Summary', 'Enunciation','Tone','Genre','Speech standard', 'Literary form',
                   'Active character','Intertextuality', 'Trope','Narrative arc'], 
    
    'frequency': [10, 10, 10, 10, 10, 9, 5, 3,4,5]
}

# Creating a DataFrame
df = pd.DataFrame(data)

# Sorting the DataFrame by frequency in descending order
df_sorted = df.sort_values(by='frequency', ascending=False)

# Creating a horizontal bar chart
plt.figure()
bars = plt.barh(df_sorted['annotation'], df_sorted['frequency'])
plt.xlabel('Frequency')
plt.ylabel('Annotation')

# Adding the frequency numbers on the bars
for bar in bars:
    plt.text(bar.get_width(), bar.get_y() + bar.get_height()/2, 
             f'{bar.get_width():.0f}',
             va='center', ha='left')

# Invert y axis to have the largest at the top
plt.gca().invert_yaxis() 
plt.show()
```
