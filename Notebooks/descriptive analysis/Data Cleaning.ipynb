## Data Preprocessing

This notebook contains the codes of data cleaning on the 50 English excerpts for descriptive analysis.

```python
import pandas as pd
```

```python
df = pd.read_csv('C:/users/idaid/Desktop/50 excerpts_descriptive_analysis.csv')
df
```

```python
import re

def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-z\s]', '', text)  
    text = re.sub(r'\s+', ' ', text).strip() 
    return text

df['full_text'] = df['full_text'].apply(clean_text)
df
```

```python
df.to_csv('C:/users/idaid/Desktop/cleaned_50_excerpts_descriptive_analysis.csv', index=False)
```
